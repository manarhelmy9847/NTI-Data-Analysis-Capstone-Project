#**Data Quality Issues**

Droping Dupilcated Rows

```python
data.drop_duplicates(inplace=True)
```

Get the total null values in each column

```python
data.isnull().sum()
```

*   Rating Column values should be between 0 and 10
*   Any Value bigger than 10 will be multiplied by (0.1) to get the exact rating




```python
data['Rating'] = np.where(data['Rating'] >10, data['Rating'] * 0.1, data['Rating'])
```

Remove any characters from the Unit Price Column values -> after this Line , the Unit Price Column will only have numerical values without any characters

```python
data['Unit price'] = data['Unit price'].str.replace('[^\d.]', '', regex=True)
```

*   Converting all Column values to float
*   As if the column has any string in it
*   This conversion done to prepare that column for numerical operations





```python
data['Unit price'] = data['Unit price'].astype(float)
```



1.   Any -ve value in Quanitity column will be cpnverted to +ve
2.   Evaluate the 5% Tax again to remove all the null values in the 'Tax 5%' Column
3.   Calculate the Total
4.   Rounding the values in the 'Tax 5%' and 'Total' columns to 4 decimal places for precision.






```python
data['Quantity'] = np.where(data['Quantity'] < 0, data['Quantity']*-1, data['Quantity'])
data['Tax 5%']= data['Unit price']*data['Quantity']*0.05
data['Total'] = data['Unit price'] * data['Quantity']+ data['Tax 5%']
data['Tax 5%'] = data['Tax 5%'].round(4)
data['Total'] = data['Total'].round(4)
```

Ensure that 'date' column is properly formatted and can be used in date-time calculations. Coercing errors to NaT avoid having non-datetime strings that could cause issues in the future.

```python
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
```

convert a time string into a 24-hour time format (HH:MM)

```python
def convert_time(t):

    if 'PM' in t or 'AM' in t:
        return pd.to_datetime(t, format='%I - %M %p').strftime('%H:%M')
    else:
        return pd.to_datetime(t, format='%H:%M').strftime('%H:%M')
data['Time'] = data['Time'].apply(convert_time)
```

1.   Replacing value "Memberr" with the correct one "Member"

Assuming that the '-' CsutomerType corresponds to "Normal" Customer typed people
2.   Replacing value "-" with the "Normal" customerType



```python
data['Customer type'] = data['Customer type'].replace({'Memberr': 'Member'})
data['Customer type'] = data['Customer type'].replace('-', 'Normal')
```

```python
Q1 = data['Total'].quantile(0.25)
Q3 = data['Total'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data['Total'] < lower_bound) | (data['Total'] > upper_bound)]

print(outliers)
```

# **Data Tidiness Issues**

*   There is a Tidiness issue which is (Column Headers as values)
*   Columns {Yangon , Naypyitaw , Mandaly} should all be values under only one column named "City"




```python
def get_city(row):
    if row['Yangon'] == 1:
        return 'Yangon'
    elif row['Naypyitaw'] == 1:
        return 'Naypyitaw'
    elif row['Mandalay'] == 1:
        return 'Mandalay'
    return None


data['City'] = data.apply(get_city, axis=1)
```

*   Droping these columns as they are not needed now
*   Now, They are values under Column 'City'




```python
data.drop(['Yangon', 'Naypyitaw', 'Mandalay'], axis=1, inplace=True)
```

There is an Tideiness issue which is (Multiple Columns for one variable)
*   Merging the 'City' and 'Branch' Column
*   Drop each of the city and Branch Columns seperately



```python
data['City - Branch'] = data['City'] + ' - ' + data['Branch']
data.drop(['City', 'Branch'], axis=1, inplace=True)
```

Adding a Column 'Day of Week'

```python
data['Day of Week'] = data['Date'].dt.day_name()
```

Adding a Column 'Month'

```python
data['Month'] = data['Date'].dt.month_name()
```

Adding a Column 'Rating Category' with grouping any rating within 4-6 , 6-8 & 8-10

```python
def CategorizeRating(rating):
    if 4 <= rating < 6:
        return '4-6'
    elif 6 <= rating < 8:
        return '6-8'
    elif 8 <= rating <= 10:
        return '8-10'
    else:
        return 'Out of Range'

data['Rating Category'] = data['Rating'].apply(CategorizeRating)

data
```

```python
jan = data[data['Month'] == 'January']
feb = data[data['Month'] == 'February']
mar = data[data['Month'] == 'March']

plt.figure(figsize=(14, 6))
jan.groupby('Date')['Total'].sum().plot()
plt.title('Sales Trends in January')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)

plt.figure(figsize=(14, 6))
feb.groupby('Date')['Total'].sum().plot()
plt.title('Sales Trends in February')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)

plt.figure(figsize=(14, 6))
mar.groupby('Date')['Total'].sum().plot()
plt.title('Sales Trends in March')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```
